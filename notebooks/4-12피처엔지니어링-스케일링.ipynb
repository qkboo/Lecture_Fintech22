```python
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer

# scaler = MinMaxScaler()
# scaler = RobustScaler()
# scaler = StandardScaler()
scaler = Normalizer()

scaler.fit(X_train) # target 은 지정하지 않음
X_train_scaled = scaler.transform(X_train) # 변환
```

```python
plt.subplot(121)
plt.scatter(X_train[:,0], X_train[:,1],
            c=y_train)
plt.title('Before scale')
plt.subplot(122)
plt.scatter(X_train_scaled[:,0],
            X_train_scaled[:,1],
            c=y_train)
plt.title('After scaled')
plt.show()
```

```python
plt.subplot(211)
plt.boxplot(X_train)
plt.title('Befer')
plt.subplot(212)
plt.boxplot(X_train_scaled)
plt.title('After Scaled')
plt.show()
```

```python
from sklearn.svm import SVC
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)
svm = SVC( C=100)
svm.fit(X_train, y_train)
print("테스트 세트 정확도{:.2f}".format( svm.score( X_test, y_test) ))
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm.fit(X_train_scaled, y_train)
print("테스트 세트 정확도{:.2f}".format( svm.score( X_test_scaled, y_test) ))
```
